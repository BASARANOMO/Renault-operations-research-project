```julia
function calculate_cost(instance_file,sol_file)
    data = open(joinpath("..", "sujet", instance_file)) do file
        readlines(file)
    end
    instance = lire_instance(joinpath("..", "sujet", instance_file))
    instance_solved = lire_solution(instance, joinpath("..", "sujet", sol_file))
    sol_feasibility = feasibility(instance_solved)
    sol_totalcost = cost(instance_solved, verbose=true)
    return sol_feasibility, sol_totalcost
end
```

```julia
instance_file_list = ["france.csv", "espagne.csv", "maroc.csv", "europe.csv"]
sol_file_list = ["france sol.txt", "espagne sol.txt", "maroc sol.txt", "europe sol.txt"]
feasibility_list = []
cost_list = []
for i in 1:length(instance_file_list)
    instance_file = instance_file_list[i]
    sol_file = sol_file_list[i]
    sol_feasibility, sol_totalcost = calculate_cost(instance_file,sol_file)
    append!(feasibility_list,sol_feasibility)
    append!(cost_list, sol_totalcost)
end
```

```julia
for i in 1:length(instance_file_list)
    instance_file = instance_file_list[i]
    print("The solution feasibility for : ", instance_file, " is ", feasibility_list[i], '\n')
    print("Its total cost is ", cost_list[i], '\n')
end
```
